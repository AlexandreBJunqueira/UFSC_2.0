### Importando Bibliotecas

```python
import requests
import pandas as pd
```

### Puxando Informações das Campanhas por API

```python
response = requests.get("https://outreachdashboard.wmflabs.org/campaigns/statistics.json")
dados = response.json()
dados = dados['campaigns']
dados
```

```python
#Transformando as Campanhas em DataFrame
df_campanhas = pd.DataFrame(dados)

# Formatando o DataFrame
df_campanhas.index = df_campanhas['id']
df_campanhas.drop(['id'], axis = 1, inplace=True)
df_campanhas.sort_index(inplace=True)
df_campanhas
```

```python
# Estrutura das URLs: https://outreachdashboard.wmflabs.org/campaigns/{slug}/overview
urls_campanhas_overview = []
urls_campanhas_programs = []
for slug in df_campanhas['slug']:
    urls_campanhas_overview = urls_campanhas_overview + [f"https://outreachdashboard.wmflabs.org/campaigns/{slug}/overview"]

for slug in df_campanhas['slug']:
    urls_campanhas_programs = urls_campanhas_programs + [f"https://outreachdashboard.wmflabs.org/campaigns/{slug}/programs"]
```

```python
# Adicionando os URLs para o DataFrame
df_campanhas['urls_campanhas_overview'] = urls_campanhas_overview
df_campanhas['urls_campanhas_programs'] = urls_campanhas_programs
df_campanhas
```

```python
# Formato da API de cada campanha: https://outreachdashboard.wmflabs.org/campaigns/{slug}.json
apis_campanhas_overview = []
for slug in df_campanhas['slug']:
    apis_campanhas_overview = apis_campanhas_overview + [f"https://outreachdashboard.wmflabs.org/campaigns/{slug}.json"]
apis_campanhas_overview
```

```python
# Adicionando as APIs para o DataFrame
df_campanhas['apis_campanhas_overview'] = apis_campanhas_overview
df_campanhas
```

### Importando Outras Bibliotecas

```python
from bs4 import BeautifulSoup
from concurrent.futures import ThreadPoolExecutor, as_completed
from urllib.parse import urljoin
```

```python
# Função para extrair os links de cursos
def extrair_links_cursos(url):
    links_cursos = set()

    while url:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            course_elements = soup.find_all('a', class_='course-link')
            for course_element in course_elements:
                link = urljoin(url, course_element['href'])  # Para adicionar a terminação ao link completo
                links_cursos.add(link)

            # Verificando se há outras páginas
            pagination_div = soup.find('div', class_='pagination')
            if pagination_div:
                next_page = pagination_div.find('a', class_='next_page')
                if next_page:
                    url = urljoin(url, next_page['href'])
                else:
                    url = None
            else:
                url = None
        else:
            return None
    return list(links_cursos)
```

```python
with ThreadPoolExecutor(max_workers=20) as executor:
    futures = []
    cursos_url = []
    for link in df_campanhas['urls_campanhas_programs']:
        futures = futures + [executor.submit(extrair_links_cursos, link)]
    for future in as_completed(futures):
        cursos_url = cursos_url + [future.result()]

# Colocando os links no DataFrame
df_campanhas['url_cursos'] = cursos_url
df_campanhas
```

```python
df_url_cursos = df_campanhas.explode('url_cursos')
df_url_cursos = df_url_cursos[['url_cursos']]
df_campanhas.drop(['url_cursos'], axis=1)  #Retirando de df_campanhas

df_url_cursos = df_url_cursos.dropna()  #Retirando valores nulos
df_url_cursos = df_url_cursos.drop_duplicates('url_cursos')  #Retirando Duplicatas

df_url_cursos = df_url_cursos.reset_index()  #Resetando o indice
df_url_cursos.drop(['id'], axis=1, inplace=True)

df_url_cursos
```

### Declarando Funções

. buscar_campanha: pega informações da campanha de um curso
. buscar_curso: pega informações de um curso
. buscar_editores: pega informações dos editores de um curso

```python
def buscar_campanha(link):
    try:
        api_campanha = f"{link}/campaigns.json"
        response_campanha = requests.get(api_campanha)
        
        if response_campanha.status_code == 200:
            campanha_data = response_campanha.json()
            if 'course' in campanha_data and 'campaigns' in campanha_data['course']:
                campaigns = campanha_data['course']['campaigns']
                if campaigns:
                    return {'Campanha': campaigns[0]['title']}  # Retorna o título da primeira campanha
        return None
    except:
        return None


def buscar_curso(link):
    try:
        api_curso = f"{link}/course.json"
        response_curso = requests.get(api_curso)
        
        if response_curso.status_code == 200:
            curso_data = response_curso.json()['course']
            return {
                'Curso': curso_data['title'],
                'Descrição': curso_data['description'],
                'Início': curso_data['start'],
                'Encerramento': curso_data['end'],
                'Instituição': curso_data['school'],
                'Link': link
            }
        
        return None
    except:
        return None


def buscar_usuarios(link):
    try:
        api_users = f"{link}/users.json"
        response_users = requests.get(api_users)
        
        if response_users.status_code == 200:
            users_data = response_users.json()
            users = {'names': [], 'roles': []}
            
            for user in users_data['course']['users']:
                if 'real_name' in user:

                    if user['real_name'] is None or user['real_name'].lower() == "null" or user['real_name'] == '':
                        users['names'] += [user['username']]
                    else:
                        users['names'] += [user['real_name']]
                    users['roles'] += [user['role']]

            return users
        return None
    except:
        return None
    
```

### Utilizando as Funções

. uso do ThreadPoolExecutor para tornar o processo mais ágil

```python
with ThreadPoolExecutor(max_workers=50) as executor:
    campanhas = list(executor.map(buscar_campanha, df_url_cursos['url_cursos']))
campanhas
```

```python
with ThreadPoolExecutor(max_workers=50) as executor:   
    cursos = list(executor.map(buscar_curso, df_url_cursos['url_cursos']))
cursos
```

```python
with ThreadPoolExecutor(max_workers=50) as executor:    
    usuarios = list(executor.map(buscar_usuarios, df_url_cursos['url_cursos']))
usuarios
```

### Organizando as Informações em uma Tabela

```python
info_cursos = []
for link, campanha, curso, usuario in zip(df_url_cursos['url_cursos'], campanhas, cursos, usuarios):
    curso_info = {'Link': link}
    if campanha:
        curso_info.update(campanha)
    info_cursos.append(curso_info)
    if curso:
        curso_info.update(curso)
    info_cursos.append(curso_info)
    if usuario:
        curso_info.update(usuario)
    info_cursos.append(curso_info)

df = pd.DataFrame(info_cursos)
df
```

### Organizando a Tabela

```python
# Definindo Função
import regex as re

def mudar(string):
    return re.sub("1", "Facilitadores", string)
```

```python
# Removando caracteres não aceitos pelo excel
df = df.applymap(lambda x: x.replace('\u000b', '') if isinstance(x, str) else x)

# Formatando as datas e ordenando dos mais recentes para mais antigos
df['Encerramento'] = pd.to_datetime(df['Encerramento'], format='%Y-%m-%dT%H:%M:%S.%fZ', errors='coerce')
df['Início'] = pd.to_datetime(df['Início'], format='%Y-%m-%dT%H:%M:%S.%fZ', errors='coerce')
df = df.dropna(subset=['Encerramento'])
df = df.sort_values(by='Encerramento', ascending=False)
df['Encerramento'] = df['Encerramento'].dt.strftime('%d/%m/%Y')
df['Início'] = df['Início'].dt.strftime('%d/%m/%Y')

# Corrigindo os nomes
def converter_lista_para_str(valor):
    if isinstance(valor, list):
        return ', '.join(map(str, valor)) 
    return valor

df = df.applymap(converter_lista_para_str)

# Reordenando as colunas
df.rename(columns={'names': 'Nome/Username'}, inplace=True)
df.rename(columns={'roles': 'Role/Posição'}, inplace=True)
ordem = ['Campanha', 'Curso', 'Descrição', 'Instituição','Início', 'Encerramento', 'Nome/Username', 'Role/Posição', 'Link']
df = df[ordem]
df = df.drop_duplicates(subset=df.columns)
df['Role/Posição'] = df['Role/Posição'].apply(mudar)
df.reset_index(drop=True, inplace=True)
df
```
